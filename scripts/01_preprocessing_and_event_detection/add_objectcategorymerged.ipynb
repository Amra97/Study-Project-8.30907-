```python
import os
import numpy as np
import pandas as pd
```

```python
CSV = r"C:\Users\Marcel\Desktop\Study Project\VR Data\Data with Turns\combined_dataframe.csv"

def merge_cat(val):
    if pd.isna(val):
        return np.nan
    c = str(val).strip()
    if c in ("Active_Agent_Face", "Passive_Agent_Face"):
        return "Face"
    if c in ("Active_Agent", "Passive_Agent"):
        return "Body"
    if c in ("Building", "TaskBuilding_Public", "TaskBuilding_Residential"):
        return "Building"
    if c == "Global_Landmark":
        return "Landmark"
    if c == "Background":
        return "Background"
    return "Other"

# ====== Load, transform, save ======
df = pd.read_csv(CSV)

df["ObjectCategoryMerged"] = df["categorical_names"].apply(merge_cat)

df["ObjectCategoryMerged"] = pd.Categorical(df["ObjectCategoryMerged"], ordered=False)
```

```python
df.to_csv(CSV, index=False)
```
